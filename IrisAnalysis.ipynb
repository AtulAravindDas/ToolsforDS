# Iris Flower Analysis with Pandas

This notebook analyzes the Iris Flower Dataset and creates valuable visualizations

## Load the dataset

```python
import pandas as pd
import seaborn as sns
```

```python
iris_df=sns.load_dataset('iris')
```

## Data loading and Exploration


```python
# Display the shape and the columns of the dataset
print("Dataset Shape:",iris_df.shape)
print(f"\nColumns:{iris_df.columns.tolist()}")

#Display the first few rows of the dataset
print("\nFirst 5 rows of the Iris dataset:\n")
print(iris_df.head())

#Get basic informatio about the dataset
print("\nIris Dataset Information:\n")
print(iris_df.info())

#Get summary statistics of the dataset
print("\nSummary Statistics of the Iris Dataset:\n")
print(iris_df.describe())
```

## Basic Data Manipulation

```python
# Create petal area column
iris_df['petal_area'] = iris_df['petal_length'] * iris_df['petal_width']

# Create sepal area column
iris_df['sepal_area'] = iris_df['sepal_length']*iris_df['sepal_width']

# Create size category column
def categorize_size(petal_area):
    if petal_area<2 :
        return 'Small'
    elif petal_area>=2 and petal_area<=5:
        return 'Medium'
    else:
        return 'Large'

iris_df['size_category'] = iris_df['petal_area'].apply(categorize_size)

print("New columns created:")
print(iris_df[['petal_length', 'petal_width', 'petal_area', 'size_category']].head(10))
```

### Alternative method to do the above

```python
iris_df_sizes = iris_df['petal_area'].apply(lambda x: 'Small' if x<2 else ('Medium' if 2<=x<=5 else 'Large'))
print(iris_df_sizes.head(10))
```

## Data Filtering

```python
# Filtering the dataset for 'setosa' species
setosa_flowers= iris_df[iris_df['species']=='setosa']
print(f"Setosa Species Count in the Iris Dataset: {len(setosa_flowers)} out of {len(iris_df)} flowers")

#Flitering for large flowers
large_flowers = iris_df[iris_df['size_category']=='Large']
print(f"\nLarge Flowers Count in the Iris Dataset: {len(large_flowers)} out of {len(iris_df)} flowers")

# Find flowers with sepal_length > 6
long_sepal = iris_df[iris_df['sepal_length']>6]
print(f"\nFlowers with sepal_length > 6: {len(long_sepal)} out of {len(iris_df)} flowers")

```

## Basic Analysis

```python
species_counts = iris_df['species'].value_counts() #Count the number of occurrences of each species
print("\nSpecies Counts:\n", species_counts)

#Find average petal length for each species
avg_petal_length =iris_df.groupby('species')['petal_length'].mean().reset_index(name='avg_petal_length')
print("\nAverage Petal Length by Species:\n", avg_petal_length)

#Find maximum petal width for each species
max_petal_width = iris_df.groupby('species')['petal_width'].max().reset_index(name='max_petal_width')
print("\nMaximum Petal Width by Species:\n", max_petal_width)
```

## Simple Visualization

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(2,2,figsize=(15,10))

#Histogram of petal lengths
axes[0, 0].hist(iris_df['petal_length'], bins=15, alpha=0.7, edgecolor='black', color='skyblue')
axes[0, 0].set_title('Distribution of Petal Lengths')
axes[0, 0].set_xlabel('Petal Length (cm)')
axes[0, 0].set_ylabel('Frequency')

#Bar chart of species counts
species_counts=iris_df.groupby('species').size().reset_index(name='counts')  
species_counts
axes[0, 1].bar(species_counts['species'], species_counts['counts'], color=['lightcoral', 'lightgreen', 'lightblue'])
axes[0, 1].set_title('Counts of Each Iris Species')
axes[0, 1].set_xlabel('Species')
axes[0, 1].set_ylabel('Count')

#Scatter plot for sepal length vs sepal width for every species
species_details=iris_df['species'].unique()
for species in species_details:
    subset=iris_df[iris_df['species']==species]
    axes[1, 0].scatter(subset['sepal_length'], subset['sepal_width'], label=species)
axes[1, 0].set_title('Sepal Length vs Sepal Width')
axes[1, 0].set_xlabel('Sepal Length (cm)')
axes[1, 0].set_ylabel('Sepal Width (cm)')
axes[1, 0].legend()

#Petal length vs petal width scatter plot
axes[1, 1].scatter(iris_df['petal_length'], iris_df['petal_width'], alpha=0.7, s=50)
axes[1, 1].set_title('Petal Length vs Petal Width')
axes[1, 1].set_xlabel('Petal Length (cm)')
axes[1, 1].set_ylabel('Petal Width (cm)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

```python
size_categories = iris_df['size_category'].value_counts()
size_categories.plot(kind='bar', color=['lightblue', 'lightgreen', 'salmon'],alpha=1, edgecolor='black', figsize=(8,5))

plt.title('Counts of Each Size Category')
plt.xlabel('Size Category')
plt.ylabel('Count')
plt.show()
```

### Using Groupby Operations


```python
# Average petal length by species using groupby
print("Average petal length by species:")
petal_by_species = iris_df.groupby('species')['petal_length'].agg(['mean', 'std', 'count'])
petal_by_species.columns = ['Average Petal Length', 'Std Dev', 'Count']
display(petal_by_species)

# Count of flowers by species using groupby
print("\nFlowers by species:")
species_counts = iris_df.groupby('species').size()
print(species_counts)

# Average measurements by species
print("\nAverage measurements by species:")
measurements_by_species = iris_df.groupby('species')[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].mean()
display(measurements_by_species.round(2))
```

### Correlation Heatmap


```python
# Create correlation heatmap for numerical columns
import numpy as np
plt.figure(figsize=(8, 6))
numerical_cols = iris_df.select_dtypes(include=[np.number])
correlation_matrix = numerical_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix of Iris Measurements')
plt.tight_layout()
plt.show()
```

## Key Insights and Findings

### Summary of Analysis Results


```python
# Generate summary insights
print("=== IRIS FLOWER ANALYSIS SUMMARY ===\n")

print(f"Dataset Overview:")
print(f"- Total flowers analyzed: {len(iris_df)}")
print(f"- Number of species: {iris_df['species'].nunique()}")
print(f"- Species: {', '.join(iris_df['species'].unique())}")
print(f"- Average petal length: {iris_df['petal_length'].mean():.2f} cm")
print(f"- Average sepal length: {iris_df['sepal_length'].mean():.2f} cm")

print(f"\nSpecies Distribution:")
species_counts = iris_df['species'].value_counts()
for species, count in species_counts.items():
    percentage = count / len(iris_df) * 100
    print(f"- {species}: {count} flowers ({percentage:.1f}%)")

print(f"\nKey Findings:")
print(f"- Setosa has the smallest petals (avg: {iris_df[iris_df['species'] == 'setosa']['petal_length'].mean():.2f} cm)")
print(f"- Virginica has the largest petals (avg: {iris_df[iris_df['species'] == 'virginica']['petal_length'].mean():.2f} cm)")
print(f"- Petal length and width are highly correlated: {iris_df['petal_length'].corr(iris_df['petal_width']):.3f}")
print(f"- Sepal length and width are moderately correlated: {iris_df['sepal_length'].corr(iris_df['sepal_width']):.3f}")

print(f"\nTop 3 largest flowers (by petal area):")
# Create petal area if not already created
if 'petal_area' not in iris_df.columns:
    iris_df['petal_area'] = iris_df['petal_length'] * iris_df['petal_width']

top_flowers = iris_df.nlargest(3, 'petal_area')[['species', 'petal_area', 'petal_length', 'petal_width']]
for idx, row in top_flowers.iterrows():
    print(f"- {row['species']}: area={row['petal_area']:.2f}, length={row['petal_length']:.2f}, width={row['petal_width']:.2f}")
```

## Statistic Summary by Species

```python
# Detailed statistical summary by species
print("=== DETAILED STATISTICS BY SPECIES ===\n")

for species in iris_df['species'].unique():
    species_data = iris_df[iris_df['species'] == species]
    print(f"{species.upper()} SPECIES:")
    print(f"  Count: {len(species_data)} flowers")
    print(f"  Petal Length: {species_data['petal_length'].mean():.2f}  {species_data['petal_length'].std():.2f} cm")
    print(f"  Petal Width: {species_data['petal_width'].mean():.2f}  {species_data['petal_width'].std():.2f} cm")
    print(f"  Sepal Length: {species_data['sepal_length'].mean():.2f}  {species_data['sepal_length'].std():.2f} cm")
    print(f"  Sepal Width: {species_data['sepal_width'].mean():.2f}  {species_data['sepal_width'].std():.2f} cm")
    if 'petal_area' in iris_df.columns:
        print(f"  Average Petal Area: {species_data['petal_area'].mean():.2f} cm")
    print()
```
